3. Ejecutamos **hadoop jar** con el fin de correr uno de los ejemplos por defecto, en este caso el *grep* que busca expresiones regulares dentro de los ficheros que le especifiquemos.

- **/usr/lib/hadoop/bin/hadoop** Es el directorio donde esta el ejecutable de hadoop en el sistema.
- **jar** Le indica a hadoop que deseamos ejecutar una aplicacion empaquetada de Java. (Jar)
- **/usr/lib/hadoop/share/hadoop/mapreduce/hadoop-mapreduce-examples-3.3.6.jar** Es la ruta donde esta el Jar que deseamos ejecutar. La versión del jar depende de la versión de hadoop instalada.

- **grep** Es un *parámetro* de los muchos que se le pueden pasar al Jar de ejemplos que trae Hadoop. grep sirve para encontrar y contar ocurrencias de strings haciendo uso de expresiones regulares.

- **input** El directorio de entrada. Es donde el programa va a buscar los archivos de entrada a la tarea de map-reduce. Aquí copiamos unos archivos de prueba en un comando anterior.
- **grep_example** El directorio de salida. Es donde el programa va a escribir el resultado de la ejecución de la aplicación. En este caso, la cantidad de veces que la palabra del parámetro siguiente, aparece en los archivos de entrada.
- **‘allowed[.]*’** Es la expresión regular que deseamos buscar.

```
%%bash
hadoop jar \
  /usr/lib/hadoop-3.3.6/share/hadoop/mapreduce/hadoop-mapreduce-examples-3.3.6.jar \
  grep input grep_example 'allowed[.]*'
```

Los resultados se almacenan en el directorio de salida (grep_example/) y se pueden verificar ejecutando *cat* en el directorio de salida:

```
!hdfs dfs -cat grep_example/*
```
